# Introduction to Seaborn

## 1.1 Introduction
Seaborn is a powerful Python library for creating data visualizations. 

### 1.1.1 Advantages of Seaborn
1.  Seaborn's main purpose is to make data visualization easy. It was built to automatically handle a lot of complexity behind the scenes. 
2. Seaborn works extremely well with pandas data structures. 
3. Seaborn is  built on top of Matplotlib, which is another Python visualization library. Matplotlib is extremely flexible. Seaborn allows you to take advantage of this flexibility when you need it, while avoiding the complexity that Matplotlib's flexibility can introduce.

## 1.2 Adding a third variable with hue
 Seaborn offers: the ability to quickly add a third variable to your plots by adding color.
TIPS - This dataset contains one row for each table served at a restaurant and has information about things like 

1. the bill amount, 

2. how many people were at the table, and

3.  when the table was served. 

Let's explore the relationship between the "total_bill" and "tip" columns using a scatter plot.

```python
tips = sns.load_dataset('tips')
tips.head()
```

The total bill per table (in dollars) is on the x-axis, and the total tip (in dollars) is on the y-axis. We can see from this plot that larger bills are associated with larger tips. 

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

What if we want to see which of the data points are smokers versus non-smokers?
You can set the "hue" parameter equal to the DataFrame column name "smoker" and then Seaborn will automatically color each point by whether they are a smoker. Plus, it will add a legend to the plot automatically! If you don't want to use pandas, you can set it equal to a list of values instead of a column name.

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker',data=tips)
plt.show()
```

### 1.2.1 Setting the hue order
Hue also allows you to assert more control over the ordering and coloring of each value. The "hue order" parameter takes in a list of values and will set the order of the values in the plot accordingly. Notice how the legend for smoker now lists "yes" before "no".

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker',hue_order=['Yes','No'],  data=tips)
plt.show()
```

### 1.2.2 Specifying hue colors
You can also control the colors assigned to each value using the "palette" parameter. This parameter takes in a dictionary, which is a data structure that has key-value pairs. This dictionary should map the variable values to the colors you want to represent the value. Here, we create a dictionary called "hue colors" that maps the value "Yes" to the color black and the value "No" to the color red. When we set hue equal to "smoker" and the palette parameter equal to this dictionary, we have a scatter plot where smokers are represented with black dots and non-smokers are represented with red dots.


```python
hue_colors = {'Yes':'Black', 'No':'Red'}
sns.scatterplot(x='total_bill',y='tip', data=tips, hue='smoker', palette=hue_colors)
plt.show()
```

### 1.2.3 Hue with countplots
As a final note, hue is available in most of Seaborn's plot types. For example, this count plot shows the number of observations we have for smokers versus non-smokers, and setting "hue" equal to "sex" divides these bars into subgroups of males versus females. From this plot, we can see that males outnumber females among both smokers and non-smokers in this dataset.

```python
tips.head()
sns.countplot(x='smoker', data=tips, hue='sex')
plt.show()
```

```python
# The "school" column indicates the initials of which school the student attended - either "GP" or "MS".
# How many students live in urban vs. rural areas, and does this vary based on what school the student attends? 
# Let's make a count plot with subgroups to find out.


# Fill in the palette_colors dictionary to map the "Rural" location value to the color "green" and the "Urban" 
# location value to the color "blue".
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x="school", data=students_data, hue="location", palette=palette_colors)

# Display plot
plt.show()
```

### 1.2.4 Hue with Scatter plots

```python
# Hue and scatter plots
# Create a scatter plot of absences vs. final grade
sns.scatterplot(x='absences', y='G3',data = students_data, hue='location')
# Show plot
plt.show()
```

```python
# Make "Rural" appear before "Urban" in the plot legend.
sns.scatterplot(x='absences', y='G3',data = students_data, hue='location', hue_order=['Rural','Urban'])
plt.show()
```
