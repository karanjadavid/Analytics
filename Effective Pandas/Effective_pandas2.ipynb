### Chain

Chaining is also called flow programming. Rather than making intermediate variables, just leverage the fact that most operations return a new object and work on that.

The chain should read like a recipe of ordered steps

(Hint: Leverage .pipe if you can't find a way to pipe)

```python
# Easy to debug
# - assign to var (df3)
# - comment out
# - pipe to display

def get_var(df, var_name):
    globals()[var_name] = df
    return df

def tweak_autos(autos):
    return (autos
            [cols]
            # create var
            .pipe(get_var, 'df3')
            .assign(cylinders = autos.cylinders.fillna(0).astype('int8'),
                   displ = autos.displ.fillna(0).astype('float16'),
                   drive = autos.drive.fillna('Other').astype('category'),
                   automatic = autos.trany.str.contains('Auto'), # indicator true/false
                   speeds = autos.trany.str.extract(r'(\d)+').fillna('20').astype('int8'), # extract the digits
                   createdOn = pd.to_datetime(autos.createdOn.replace({'EDT':'-04:00', 'EST':'-05:00'},regex=True)),
                   ffs = autos.eng_dscr.str.contains('FFS')
                   )
            # debug pipe
            .pipe(lambda df:display(df) or df)
            .astype({'highway08':'int8', 'city08':'int16', 'comb08':'int16', 'fuelCost08':'int16', 'range':'int16', 'year':'int16', 'make':'category'})
            .drop(columns = ['trany', 'eng_dscr'])
            )

tweak_autos(autos)
```

The function get_var takes two parameters:

df: A pandas DataFrame object.

var_name: A string representing the name of a variable.

The purpose of the function is to assign the DataFrame object df to a variable with the name specified by var_name in the global namespace, so that it can be accessed outside the function.

The function achieves this by calling the built-in globals() function to obtain a reference to the global namespace, and then assigning the DataFrame object df to a new variable with the name specified by var_name using square bracket notation ([]).

Finally, the function returns the original DataFrame object df.

Note that this function can be potentially dangerous because it modifies the global namespace, which can lead to unintended consequences and make the code harder to reason about. It is generally recommended to avoid modifying the global namespace unless absolutely necessary.

```python
def country(val):
    if val in {'Chevrolet', 'Ford', 'Dodge', 'GMC', 'Tesla'}:
        return 'US'
    return 'Other'

# add country
(autos2
 .assign(country=autos.make.apply(country))
 .groupby(['year','country'])
 .mean()
)
```

```python
# applying multiple aggregates
def second_to_last(ser):
    return ser.iloc[-2]

(autos2
 .assign(country = autos2.make.apply(country))
 .groupby(['year','country'])
 .agg(['min', 'max', second_to_last])
)
```

unstack( ) Take the innermost index which is country and rotate it. Stick it into the columns. We make hierchical columns. 

```python
(autos2
 .assign(country=autos2.make.apply(country))
 .groupby(['year','country'])
 .mean()
 .unstack()
)
```

```python
(autos2
 .assign(country=autos2.make.apply(country))
 .groupby(['year','country'])
 .mean()
 .unstack()
 .city08 # pull out city08 column
 .plot()
 .legend(bbox_to_anchor=(1,1))
)
```

```python
# smoothe it out abit w/ rolling
(autos2
 .assign(country=autos2.make.apply(country))
 .groupby(['year','country'])
 .mean()
 .unstack()
 .rolling(2) # smooth out the plot
 .city08 # pull out city08 column
 .mean()
 .plot()
 .legend(bbox_to_anchor=(1,1))
)
```

```python
# filter after grouping
def vals_gt(df_, num): # take a dataframe and number
    return df_[df_.gt(num)].dropna() # return the dataframe where the dataframe.gt is greater than number. Drop NAs

(autos2
 .assign(country = autos2.make.apply(country))
 .groupby(['year', 'country'])
 .count()
 .pipe(vals_gt, 700)
)
```
